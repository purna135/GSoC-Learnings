## Calculating memory size of diff data

```python
from sys import getsizeof

g = globals()
data = []

for i in range(1,16):
    data_name = 'data{}'.format(i)
    data_nd = g[data_name].ndim
    data_size = g[data_name].shape
    data_MB = round(getsizeof(g[data_name]) / 1024 / 1024, 4)
    data.append([data_nd, data_size, data_MB])
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['N-dims', 'Shape of Data', 'Memory Size(in MB)'])
df
```

## Runtime of Aesara-Scipy with Vectorization

```python
x = at.tensor3()
f = aesara.function([x], at.linalg.cholesky(x))
```

```python
scipy_runtime = pd.DataFrame()
```

```python
from IPython.display import clear_output
```

```python
g = globals()

for i in range(1,16):
    clear_output(wait=True)
    run_time = []
    data = g['data{}'.format(i)]
    
    for iteration in range(30):
        time = %timeit -o f(data)
        run_time.append(time.average)
    
    scipy_runtime['data{}'.format(i)] = run_time
```

```python
scipy_runtime
```

### Saving result to csv file

```python
scipy_runtime.to_csv("Aesara_Scipy_3d.csv", index = False)
```

```python
scipy_runtime.mean(axis=0)
```

```python
df["Scipy_AvgTime"] = scipy_runtime.mean(axis=0).values
df
```

## Runtime of Aesara-Numpy implementation

```python
x = at.tensor3()
f_np = aesara.function([x], at.linalg.cholesky_np(x))
```

```python
numpy_runtime = pd.DataFrame()
```

```python
g = globals()

for i in range(1,16):
    clear_output(wait=True)
    run_time = []
    data = g['data{}'.format(i)]
    
    for iteration in range(30):
        time = %timeit -o f_np(data)
        run_time.append(time.average)
    
    numpy_runtime['data{}'.format(i)] = run_time
```

```python
numpy_runtime
```

### Saving result to csv file

```python
numpy_runtime.to_csv("Aesara_Numpy_3d.csv", index = False)
```

```python
df["Numpy_AvgTime"] = numpy_runtime.mean(axis=0).values
df
```

```python
df.to_csv("Cholesky_test_3d.csv", index = False)
```

## Visualizing Runtime

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Numpy_AvgTime"], lw = 2, marker = 'o', ms = 8, label = "Aesara Numpy Runtime")
plt.plot(df["Scipy_AvgTime"], lw = 2, marker = 'd', ms = 8, label = "Aesara Scipy Runtime")
plt.ylabel("Runtime in Sec", fontsize=18)
plt.legend()
```

```python
plt.figure(figsize=(12,6))

plt.bar(range(15), df["Memory Size(in MB)"]/80, width = 0.4, color='c') #/80 to reduce height of bargraph
plt.plot(df["Numpy_AvgTime"], lw = 3, marker = 'o', ms = 10, label = "Aesara Numpy Runtime")
plt.plot(df["Scipy_AvgTime"], lw = 3, marker = 'd', ms = 10, label = "Aesara Scipy Runtime")

plt.xlabel("Data Size in MB", fontsize=18)
plt.ylabel("Runtime in Sec", fontsize=18)
plt.legend()
```
